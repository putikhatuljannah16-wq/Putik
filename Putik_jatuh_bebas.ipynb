```python
import matplotlib.pyplot as plt

# Parameter
g = 9.8
h0 = 100  # ketinggian awal
t = list(range(11))  # 0–10 detik

# Perhitungan
s = [h0 - 0.5 * g * (ti**2) for ti in t]   # posisi
v = [g * ti for ti in t]                   # kecepatan
a = [g for _ in t]                         # percepatan

# Grafik
plt.plot(t, s, label="Posisi (m)", marker='o')
plt.plot(t, v, label="Kecepatan (m/s)", marker='s')
plt.plot(t, a, label="Percepatan (m/s²)", marker='^')
plt.title("Kasus 1: Jatuh Bebas Standar")
plt.xlabel("Waktu (s)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Maksimum posisi:", max(s), "meter")
print("Maksimum kecepatan:", max(v), "m/s")
```

```python
import numpy as np

# Parameter
g = 9.8
v0 = 10
h0 = 120
t = np.arange(0, 10.5, 0.5)  # setiap 0,5 detik

# Perhitungan
s = [h0 - (v0*ti + 0.5*g*ti**2) for ti in t]
v = [v0 + g*ti for ti in t]
a = [g for _ in t]

# Grafik
plt.plot(t, s, label="Posisi (m)", marker='o')
plt.plot(t, v, label="Kecepatan (m/s)", marker='s')
plt.plot(t, a, label="Percepatan (m/s²)", marker='^')
plt.title("Kasus 2: Jatuh Bebas dengan v0")
plt.xlabel("Waktu (s)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Maksimum posisi:", max(s), "meter")
print("Maksimum kecepatan:", max(v), "m/s")
```
